```python
import pandas as pd
```

```python
demographics_df = pd.read_csv('Demographics_Data.csv', index_col=0)
blood_pressure_df = pd.read_csv('Blood_Pressure_Examination_Data.csv', index_col=0)
body_measures_df = pd.read_csv('Body_Measures_Examination_Data.csv', index_col=0)
alcohol_df = pd.read_csv('Alcohol_Use_Questionnaire_Data.csv', index_col=0)
blood_pressure_cholestrol_df = pd.read_csv('Blood_Pressure_Cholestrol_Questionnaire_Data.csv', index_col=0)
consumer_behavior_df = pd.read_csv('Consumer_Behavior_Questionnaire_Data.csv', index_col=0)
diabetes_df = pd.read_csv('Diabetes_Questionnaire_Data.csv', index_col=0)
diet_behavior_nutrition_df = pd.read_csv('Diet_Behavior_Nutrition_Questionnaire_Data.csv', index_col=0)
medical_conditions_df = pd.read_csv('Medical_Conditions_Questionnaire_Data.csv', index_col=0)
physical_activity_df = pd.read_csv('Physical_Activity_Questionnaire_Data.csv', index_col=0)
smoking_df = pd.read_csv('Smoking_Questionnaire_Data.csv', index_col=0)
target_df = pd.read_csv('Target.csv', index_col=0)
```

```python
print(demographics_df.shape)
print(blood_pressure_df.shape)
print(body_measures_df.shape)
print(alcohol_df.shape)
print(blood_pressure_cholestrol_df.shape)
print(consumer_behavior_df.shape)
print(diabetes_df.shape)
print(diet_behavior_nutrition_df.shape)
print(medical_conditions_df.shape)
print(physical_activity_df.shape)
print(smoking_df.shape)
print(target_df.shape)
```

```python
master_raw = pd.merge(demographics_df, blood_pressure_df, on='SEQN', how='left')
master_raw = pd.merge(master_raw, body_measures_df, on='SEQN', how='left')
master_raw = pd.merge(master_raw, alcohol_df, on='SEQN', how='left')
master_raw = pd.merge(master_raw, blood_pressure_cholestrol_df, on='SEQN', how='left')
master_raw = pd.merge(master_raw, consumer_behavior_df, on='SEQN', how='left')
master_raw = pd.merge(master_raw, diabetes_df, on='SEQN', how='left')
master_raw = pd.merge(master_raw, diet_behavior_nutrition_df, on='SEQN', how='left')
master_raw = pd.merge(master_raw, medical_conditions_df, on='SEQN', how='left')
master_raw = pd.merge(master_raw, physical_activity_df, on='SEQN', how='left')
master_raw = pd.merge(master_raw, smoking_df, on='SEQN', how='left')
master_raw = pd.merge(master_raw, target_df, on='SEQN', how='left')
```

```python
master_data = master_raw[(master_raw['Age']>=18) & (master_raw['Age']<66)]
```
