```python
df_features.drop(columns='SEQN', inplace=True)
```

```python
categorical_variables = ['Gender', 'Race', 'Country_Born', 'Family_Income', 'Pulse_Regular_Irregular', 
                         'Ever_told_high_blood_pressure', 'Ever_told_high_cholesterol', 
                         'Doctor_told_you_have_diabetes', 'Past_30day_milk_consumption', 'Doctor_told_you_overweight', 
                         'Doctor_told_you_lose_weight', 'Doctor_told_you_exercise', 'Doctor_told_you_reduce_salt', 
                         'Doctor_told_you_reduce_fat_calories', 'Walk_or_bicycle', 'Vigorous_recreational_activities', 
                         'Moderate_recreational_activities', 'Smoked_at_least_100_cigarettes_in_life']
```

```python
df_features = pd.get_dummies(df_features, columns=categorical_variables)
```

```python
df_target = pd.get_dummies(df_target)
```

```python
df_target.rename(columns={1.0:'Target', 2.0:'to_drop'}, inplace=True)
```

```python
df_target = df_target['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_features, df_target, test_size=0.2)
```

```python
sum(y_train==0)
```

```python
sum(y_train==1)
```

```python
sm = SMOTE()
```

```python
X_train, y_train = sm.fit_sample(X_train, y_train.ravel())
```

```python
sum(y_train==0)
```

```python
sum(y_train==1)
```

```python
model = RandomForestClassifier(n_estimators=100,max_depth=5) 
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_train, y_train)
```

```python
print(metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix

print('Accuracy Score:  ' + str(accuracy_score(y_test, y_pred)))
print('F1 Score:        ' + str(f1_score(y_test, y_pred, average="macro")))
print('Precision Score: ' + str(precision_score(y_test, y_pred, average="macro")))
print('Recall Score:    ' + str(recall_score(y_test, y_pred, average="macro")) )
report = classification_report(y_test, y_pred)
print ('')
print (report)
```
